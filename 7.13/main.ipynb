```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.cm import ScalarMappable

dt = 10
duration = 100
ts = np.linspace(0, duration, dt)
ys = np.linspace(0, duration, dt)
a = 10
k = 1e-4
fluxs = np.full((duration, dt), np.nan)
Bs = np.full((duration,dt), np.nan)
j = 0

for t in ts:
    i = 0
    flux_value = (1/4)*k*t**2*a**5
    fluxs[j] = flux_value

    for y in ys:
        B_value = k * y**3 * t**2
        Bs[j][i] = B_value
        i = i + 1

    plot = plt.plot(ys, Bs[j])
    plt.show()
    plt.ylim(0, 1e6)
    plt.xlim(0, duration)
    plt.xlabel('y')
    plt.ylabel('Magnetic field')
    plt.title(f'Magnetic field vs. y-position at t = {round(t, 3)}')
    
    j = j + 1

print(fluxs)
print(Bs)

plt.figure(size=(10,8))
plot = plt.plot(ts, fluxs)
plt.xlim(0, duration)
plt.show()
    
```
